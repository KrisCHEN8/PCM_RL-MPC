```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import warnings
from torch import nn
import os
from numpy.polynomial import Polynomial
import cvxpy as cp
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.linear_model import BayesianRidge, ARDRegression
from sklearn.ensemble import GradientBoostingRegressor
from sysidentpy.model_structure_selection import MetaMSS, FROLS
from sysidentpy.metrics import root_relative_squared_error
from sysidentpy.basis_function._basis_function import Polynomial, Fourier
from sysidentpy.utils.display_results import results
from sysidentpy.utils.plotting import plot_residues_correlation, plot_results
from sysidentpy.residues.residues_correlation import (
    compute_residues_autocorrelation,
    compute_cross_correlation,
)
from sysidentpy.general_estimators import NARX
from catboost import CatBoostRegressor
from sysidentpy.utils.save_load import save_model, load_model

warnings.filterwarnings('ignore')
plt.rcParams["font.family"] = "Times New Roman"
```

```python
# read data from pickle (you may need to change the path for pickled data)
building_results = pd.read_pickle('C:/Users/yangzhec/OneDrive - KTH/Paper with CNR/data/building_results.pkl')
office_elec = pd.read_pickle('C:/Users/yangzhec/OneDrive - KTH/Paper with CNR/data/office_load.pkl')
```

```python
# calculate the average temp by assuming single thermal zone
building_results['avetemp'] = (building_results['TAIR_Office1             '] + building_results['TAIR_Office2             '] +
                             building_results['TAIR_Office3             '] + building_results['TAIR_Office4             '] +
                             building_results['TAIR_Office5             '] + building_results['TAIR_Office6             '] +
                             building_results['TAIR_Office7             '] + building_results['TAIR_Office8             ']) / 8
```

```python
# seperate cooling season data (Jun, Jul, and Aug) 
office_elec_cs = office_elec[(office_elec.index.month==6) | (office_elec.index.month==7) | (office_elec.index.month==8)]
building_results_cs = building_results[(building_results.index.month==6) | (building_results.index.month==7) | (building_results.index.month==8)]
```

```python
total_df = pd.concat([building_results_cs, office_elec_cs], axis=1)
```

```python
# split training and validation sets (Jun Jul as training, Aug as validation)
train_df = total_df[total_df.index.month!=8]
test_df = total_df[total_df.index.month==8]

X_train = train_df[['Electricity Demand-TOT Cooling', 'T                        ']].values.astype(float)
y_train = train_df['avetemp'].values.reshape(-1,1).astype(float)

X_test = test_df[['Electricity Demand-TOT Cooling', 'T                        ']].values.astype(float)
y_test = test_df[['avetemp']].values.reshape(-1,1).astype(float)

print(X_train)
print('-'*50)
print(y_train)
```

# Polynomial with FROLS estimator

```python
basis_function = Polynomial(degree=2)

frols_model = FROLS(
    order_selection=False,
    n_terms=5,
    extended_least_squares=False,
    ylag=1,
    xlag=[[1], [1]],
    info_criteria="lilc",
    estimator="recursive_least_squares",
    basis_function=basis_function,
)

frols_model.fit(X=X_train, y=y_train)
```

```python
y_hat = frols_model.predict(X=X_test, y=y_test, steps_ahead=1)
rrse = root_relative_squared_error(y_test, y_hat)
print(rrse)
```

```python
r = pd.DataFrame(
    results(
        frols_model.final_model,
        frols_model.theta,
        frols_model.err,
        frols_model.n_terms,
        err_precision=8,
        dtype="sci",
    ),
    columns=["Regressors", "Parameters", "ERR"],
)
print('Detailed parameters:')
print(r)
print('-'*50)
plot_results(y=y_test, yhat=y_hat, n=800, title="NARMAX model validation",
    figsize=(10, 6),
    style='seaborn-v0_8-notebook',
    facecolor="white")
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.style.use('default')
plt.figure(figsize=(10, 6))

plt.plot(office_elec_cs[office_elec_cs.index.month==8].index[0:1345], y_hat[0:1345], label='model', lw=1.4, marker='*', markersize=7)
plt.plot(office_elec_cs[office_elec_cs.index.month==8].index[0:1345], y_test[0:1345], label='data', lw=1.4, marker='o', markersize=4)
plt.ylabel('Indoor Temperature ($^\circ$C)', fontsize=14)
plt.xlabel('Datetime', fontsize=14)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=14)
plt.title('NARMAX Model Validation', fontsize=14)

plt.tight_layout()
plt.savefig('./conference_results', dpi=500, bbox_inches='tight')
plt.show()
```

```python
print("MSE: ", mean_squared_error(y_test, y_hat))
print("MAE: ", mean_absolute_error(y_test, y_hat))
print("STD: ", (y_test - y_hat).std())
print("MAPE: ", mean_absolute_percentage_error(y_test, y_hat))
```

# Polynomial with general estimator (catboost regressor)

```python
# polynomial with catboost
basis_function = Polynomial(degree=2)

catboost_model = NARX(
    base_estimator=CatBoostRegressor(
        iterations=300,
        learning_rate=0.1,
        depth=8),
    xlag=[[1,2], [1,2]],
    ylag=2,
    basis_function=basis_function,
    model_type="NARMAX",
    fit_params={'verbose': False}
)

catboost_model.fit(X=X_train, y=y_train)
```

```python
y_hat = catboost_model.predict(X=X_test, y=y_test, steps_ahead=1)
rrse = root_relative_squared_error(y_test, y_hat)
print(rrse)
```

```python
plot_results(y=y_test, yhat=y_hat, n=1000, title="Polynomial with CatBoostRegressor validation",
    figsize=(10, 6),
    style='seaborn-v0_8-notebook',
    facecolor="white")
```

```python
print("MSE: ", mean_squared_error(y_test, y_hat))
print("MAE: ", mean_absolute_error(y_test, y_hat))
print("STD: ", (y_test - y_hat).std())
print("MAPE: ", mean_absolute_percentage_error(y_test, y_hat))
```

# Polynomial with Bayesian ridge

```python
# polynomial with catboost
basis_function = Polynomial(degree=2)

BayesianRidge_narx = NARX(
    base_estimator=BayesianRidge(),
    xlag=[[1,2], [1,2]],
    ylag=2,
    basis_function=basis_function,
    model_type="NARMAX",
)

BayesianRidge_narx.fit(X=X_train, y=y_train)
```
